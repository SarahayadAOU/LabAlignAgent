# Match and Enrich LabTest Individuals from LOINC

```python
def enrich_labtest_with_loinc(labtest, loinc_row):
    if pd.notna(loinc_row['LOINC_NUM']):
        labtest.hasLOINCCode = [str(loinc_row['LOINC_NUM'])]

    if pd.notna(loinc_row['PROPERTY']):
        labtest.hasProperty = [str(loinc_row['PROPERTY'])]

    if pd.notna(loinc_row['METHOD_TYP']):
        labtest.usesMethod = [str(loinc_row['METHOD_TYP'])]

    if pd.notna(loinc_row['EXAMPLE_UNITS']):
        labtest.hasUnit = [str(loinc_row['EXAMPLE_UNITS'])]

    if pd.notna(loinc_row['LONG_COMMON_NAME']):
        labtest.hasLongLabel = [str(loinc_row['LONG_COMMON_NAME'])]
```

```python
match_count = 0
skipped = 0

for individual in onto_kfsh.individuals():
    if hasattr(individual, "hasName") and individual.hasName:
        name = str(individual.hasName[0]).strip().lower()
        match = loinc_df[loinc_df["COMPONENT"].str.lower() == name]

        if not match.empty:
            enrich_labtest_with_loinc(individual, match.iloc[0])
            match_count += 1
        else:
            skipped += 1
    else:
        skipped += 1

print(f"✅ Enriched {match_count} LabTest individuals with LOINC metadata.")
print(f"⚠️ Skipped {skipped} individuals (no hasName or no match found).")
```

Use Fuzzy Matching Instead of Exact

```python
!pip install fuzzywuzzy[speedup]
```

```python
from fuzzywuzzy import fuzz

preview_matches = []
processed = 0

for individual in onto_kfsh.individuals():
    processed += 1

    # Extract name from the individual's IRI
    name = individual.name.strip().lower().replace("_", " ")

    best_match = None
    best_score = 0

    for _, row in loinc_df.iterrows():
        loinc_name = str(row['COMPONENT']).strip().lower()
        score = fuzz.ratio(name, loinc_name)

        if score > best_score:
            best_score = score
            best_match = row['COMPONENT']

    preview_matches.append((name, best_match, best_score))

    # Print every 10 iterations to show progress
    if processed % 10 == 0:
        print(f"Processed {processed}: {name} → {best_match} (score: {best_score})")

print(f"✅ Completed fuzzy matching for {processed} individuals.")
```

Apply Enrichment Based on Matching Results

```python
match_count = 0

for (name, best_match, score) in preview_matches:
    if score >= 70:
        # Find the ontology individual by its name
        individual = onto_kfsh.search_one(iri="*" + name.replace(" ", "_"))
        if individual:
            # Find the corresponding LOINC row
            loinc_row = loinc_df[loinc_df["COMPONENT"].str.lower() == best_match.lower()]
            if not loinc_row.empty:
                enrich_labtest_with_loinc(individual, loinc_row.iloc[0])
                match_count += 1

print(f"✅ Enriched {match_count} LabTest individuals with LOINC metadata.")
```

```python
match_count = 0

# Build a quick lookup dictionary for all individuals by lowercase name
name_to_individual = {ind.name.lower(): ind for ind in onto_kfsh.individuals()}

for (name, best_match, score) in preview_matches:
    if score >= 70:
        # Match directly using the normalized lowercase name
        ind = name_to_individual.get(name)
        if ind:
            loinc_row = loinc_df[loinc_df["COMPONENT"].str.lower() == best_match.lower()]
            if not loinc_row.empty:
                enrich_labtest_with_loinc(ind, loinc_row.iloc[0])
                match_count += 1

print(f"✅ Enriched {match_count} LabTest individuals with LOINC metadata.")
```

Now let's enrich  SGH.rdf

```python
# Step 1: Load SGH ontology
from owlready2 import *
import os

onto_sgh = get_ontology("file://" + os.getcwd() + "/sgh.rdf").load()

# Step 2: Define enrichment function
def enrich_labtest_with_loinc(labtest, loinc_row):
    if pd.notna(loinc_row['LOINC_NUM']):
        labtest.hasLOINCCode = [str(loinc_row['LOINC_NUM'])]

    if pd.notna(loinc_row['PROPERTY']):
        labtest.hasProperty = [str(loinc_row['PROPERTY'])]

    if pd.notna(loinc_row['METHOD_TYP']):
        labtest.usesMethod = [str(loinc_row['METHOD_TYP'])]

    if pd.notna(loinc_row['EXAMPLE_UNITS']):
        labtest.hasUnit = [str(loinc_row['EXAMPLE_UNITS'])]

    if pd.notna(loinc_row['LONG_COMMON_NAME']):
        labtest.hasLongLabel = [str(loinc_row['LONG_COMMON_NAME'])]

# Step 3: Fuzzy match SGH individuals to LOINC
from fuzzywuzzy import fuzz

preview_matches_sgh = []
processed = 0

print("🔍 Starting fuzzy matching...")

for individual in onto_sgh.individuals():
    processed += 1
    name = individual.name.strip().lower().replace("_", " ")
    best_match = None
    best_score = 0

    for _, row in loinc_df.iterrows():
        loinc_name = str(row['COMPONENT']).strip().lower()
        score = fuzz.ratio(name, loinc_name)

        if score > best_score:
            best_score = score
            best_match = row['COMPONENT']

    preview_matches_sgh.append((name, best_match, best_score))

    if processed % 10 == 0:
        print(f"✅ Matched {processed}: {name} → {best_match} (score: {best_score})")

print(f"\n🎯 Completed fuzzy matching for {processed} SGH individuals.\n")

# Step 4: Enrich SGH individuals with LOINC metadata
match_count = 0
enriched = 0

print("💡 Starting enrichment of individuals with LOINC data...")

name_to_individual_sgh = {ind.name.lower(): ind for ind in onto_sgh.individuals()}

for idx, (name, best_match, score) in enumerate(preview_matches_sgh):
    if score >= 70:
        ind = name_to_individual_sgh.get(name)
        if ind:
            loinc_row = loinc_df[loinc_df["COMPONENT"].str.lower() == best_match.lower()]
            if not loinc_row.empty:
                enrich_labtest_with_loinc(ind, loinc_row.iloc[0])
                enriched += 1

    if idx % 10 == 0:
        print(f"🧬 Enriched {idx} of {len(preview_matches_sgh)} processed...")

print(f"\n✅ Total enriched SGH individuals: {enriched}")

# Step 5: Save to file
onto_sgh.save(file="sgh_enriched.rdf", format="rdfxml")
print("💾 Saved as sgh_enriched.rdf")

# Step 6: Download
from google.colab import files
files.download("sgh_enriched.rdf")
```
